```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
n = np.linspace(0,N-1,N)
r = 0.87
x_o = 0.65
x_n = [x_o]


def f(r,x):
    return 4*r*x*(1-x)

for i in range(N):
    x_n.append(f(r, x_n[i]))
    
```

```python
plt.title(' $x_0 = 0.65$ and r = 0.87')
plt.plot(n,x_n[0:N], 'b--')
plt.plot(x_n[0:N], 'ro')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.show()
```

```python
xx = np.linspace(0,1,100)
x_val = xx
plt.title("The fourfold iterations has eight fixed points (intersecting the line)")
for i in range(4):
    if i !=2 :
        plt.plot(xx,f(r,x_val))
    x_val = f(r,x_val)
plt.plot([0,1],[0,1])
plt.ylabel('x')
plt.show()
```

```python
def bifurcation():
    n = 10000
    r = np.linspace(0,1,n)
    ite = 1000
    end = 100
    x = 0.00001
    for i in range(ite):
        x = f(r,x)
        if i >= (ite-end):
            plt.plot(r,x, 'k-', alpha = 0.015)
            plt.rcParams["figure.figsize"] = [16,8]

bifurcation()
plt.title(' $x = [0,1] $ and r = [0,1]')
plt.xlabel('r')
plt.axhline(0.65)
plt.axvline(0.87)
plt.ylabel('x')

plt.show()
bifurcation()
plt.title(' $x = [0,1] $ and r = [0.88,1]')
plt.xlabel('r')
plt.ylabel('x')
plt.xlim(0.88,1)
plt.show()

bifurcation()
plt.title(' $x = [0,1] $ and r = [0.95,0.97]')
plt.xlabel('r')
plt.ylabel('x')
plt.xlim(0.95,0.97)
plt.show()
```

```python
def plot(r,x,n):
    xx = np.linspace(0,1,100)
    plt.plot(xx, f(r,xx), 'r--')
    plt.plot([0,1],[0,1], 'g--')
    xlist = []     #(x     f(x)  f(x)   ff(x),...)
    ylist = []     #(f(x)  f(x)  ff(x)  ff(x),...)
    for i in range(n):     #for every i, 2 point for x list and 2 points for ylist are added
        xlist.append(x)     
        x = f(r,x)
        ylist.append(x)
        ylist.append(x)
        xlist.append(x)
    plt.plot(xlist,ylist, 'k-', alpha = 1)

plot(0.87460,0.5,5) #5 (x, f(x)) and (f(x),f(x))
plt.title('Superstable four-cycle with 'r'$r=0.874640$')
plt.xlim(0,1)
plt.ylim(0,1)
plt.rcParams["figure.figsize"] = [16,8]
plt.show()

bifurcation()
plt.xlabel('r')
plt.ylabel('x')
plt.xlim(0.85,0.90)
plt.axvline(0.87460)
plt.rcParams["figure.figsize"] = [8,8]
plt.show()

n = 100
r = 0.87464
ite = 1000
x = 0.65
x_n = []
for i in range(ite):
    x = f(r,x)
    x_n.append(x)
    
plt.hist(x_n, 100)
plt.xlim(0,1)
plt.rcParams["figure.figsize"] = [4,4]
plt.show()
```
