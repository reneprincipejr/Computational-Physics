```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logistics = lambda r,x: 4*r*x*(1-x)

x = 0.65
r = 0.87
i = range(41)
xlist = []

for ii in i:
    xlist.append(x)
    x = logistics(r,x)

plt.title('Iteration starting from 'r'$x_0 = 0.65$'' using the parameter r = 0.87. \n The 'r'$x_n$'
          ' values move towards an attractor of period 4.')
plt.xticks(range(0,41,10))
plt.xlabel('n')
plt.ylabel('x(n)')
plt.plot(i,xlist,'--bo')
```

```python
#The x_n values move towaard an attractor of period 4, which is described by the fixed points of the fourfold iterated functions
```

```python
x = np.linspace(0,1,1000)
xx = x
r = 0.87

for i in range(4):
    if i !=2 :
        plt.plot(xx,logistics(r,x),label=str(i+1))
    x = logistics(r,x)
plt.title('The function 'r'$f(x)$'' and its two- and fourfold iterations. \n' 
          'The latter has eight fixed points.')
plt.plot([0,1],[0,1],label='x = y')
plt.legend()
plt.xlabel('x')
```

```python
i = 1000
r = np.linspace(0,1,i)
x = np.linspace(0,1,i)

for ii in range(i-1): #0-999
    x = logistics(r,x)
    if ii >= 100: #not showing 0 to 99
        plt.scatter(r,x,s=0.0001)
        
plt.title('Each vertical line contains 1000 'r'$x_n$'' values from the iterated \n parabola, '
          'not showing the first 100 points. The parameter 'r'$r$'' \n varies between 0 and 1 along the horizontal.')
plt.xticks([0,0.5,1.0])
plt.xlabel('r')
plt.ylabel('x')
plt.show()  

for ii in range(i-1): #0-999
    x = logistics(r,x)
    if ii >= 100: #not showing 0 to 99
        plt.xlim(0.88,1)
        plt.scatter(r,x,s=0.004)
        
plt.title('A close-up for 'r'$r$'' between 0.88 and 1')
plt.yticks([0,0.5,1.0])
plt.xticks([0.88,0.94,1.0])
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python
def plot(r,x,n):
    xx = np.linspace(0,1,100)
    plt.plot(xx, logistics(r,xx))
    plt.plot([0,1],[0,1])
    xlist = []     #(x     f(x)  f(x)   ff(x),...)
    ylist = []     #(f(x)  f(x)  ff(x)  ff(x),...)
    for i in range(n):     #for every i, 2 point for x list and 2 points for ylist are added
        xlist.append(x)     
        x = logistics(r,x)
        ylist.append(x)
        ylist.append(x)
        xlist.append(x)
    plt.plot(xlist,ylist)

plot(0.87460,0.5,5) #5 (x, f(x)) and (f(x),f(x))
plt.title('Superstable four-cycle with 'r'$r=0.874640$')
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()

plot(0.9642,0.5,1000)
plt.title('Three chaotic bands for 'r'$r=0.9642$'', \njust above the large window with the three-cycle')
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```
