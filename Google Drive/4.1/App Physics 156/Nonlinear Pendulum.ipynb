```python
import numpy as np
from scipy.special import ellipk, ellipj
import matplotlib.pyplot as plt
from scipy.fftpack import fft

 
def T(phi0):    
    return 4*ellipk(np.sin(phi0/2)**2)

phis = np.linspace(0, np.pi) 
ts = T(phis)
int = np.pi/4

plt.plot(phis, ts, "k") 
plt.ylim(0, 20)
plt.rcParams["figure.figsize"] = [8,4]
plt.xlabel("in units of $\pi$")
plt.ylabel("T")
plt.show()
```

ellipj = to avoid divergence
ellipk = complete elliptic integral

for small phi, PE is quadratic, -> sinusoidal oscillation with period T = 2*pi(sqrt(l/g))
for larger phi, t depends on phinot



```python
 def psi(t, phi0):    
    return ellipj(t, np.sin(phi0/2)**2)[3]
def sinepsi(t, phi0):    
    return ellipj(t, np.sin(phi0/2)**2)[0]
def phinorm(x, phi0):    
    return 2*np.arcsin(np.sin(phi0/2)*sinepsi(x* T(phi0),phi0))/phi0
```

```python
phi0 = [np.pi/10, 4*np.pi/5, 19*np.pi/20, 99*np.pi/100, 999*np.pi/1000]

x = np.linspace(0,1,100)

def f(x):
    ff = []
    for i in range(5):
        ff.append(phinorm(x,phi0[i]))
    return ff

flist = np.array(f(x))

for i in range(5):
    plt.plot(x,flist[i], "k")
    plt.rcParams["figure.figsize"] = [16,8]

plt.xlabel("t/T")
plt.ylabel("phi/phi0")
plt.axhline(y=0, color = "k")
```

```python
y = phinorm(x,phi0[4])
ylist = np.abs(fft(y)/np.sqrt(len(y)))
xlist = np.arange(15)
plt.plot(xlist,ylist[:15],"o", color = "k")
plt.axhline(y=0, color = "k")
plt.rcParams["figure.figsize"] = [8,4]
plt.xlabel("s")
plt.ylabel("Abs[b(s)]")
```

This is to study the higher harmonics by breaking down the wave into the its simplest harmonic form.

```python
phi = np.linspace(-np.pi,np.pi, 1000)
phidot = np.linspace(-3,3, 1000)
E = [lambda phidot,phi: ((phidot**2)/2)-np.cos(phi)]
X, Y = np.meshgrid(phi,phidot)
Z = E[0](Y,X)
n = [-0.5, 0, 0.5, 1., 1.5, 2.]

plt.contour(X,Y,Z, levels = n)
plt.xlabel("phi")
plt.ylabel("phidot")
plt.rcParams["figure.figsize"] = [8,4]

    
```
