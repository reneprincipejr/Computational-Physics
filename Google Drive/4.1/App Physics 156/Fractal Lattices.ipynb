```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(69)
plt.rcParams["figure.figsize"] = [16,8]


def mid(p1, p2):
    return (p1 + p2) /2.0

points = 100000

#p = np.array(([0,0], [1,0],[0.5,0.5]))
p = np.array(([np.random.randint(0,50),np.random.randint(0,50)],
              [np.random.randint(0,50),np.random.randint(0,50)],
              [np.random.randint(0,50),np.random.randint(0,50)]))
q = np.array([0,0])
xlist = []
ylist = []

for n in range(points):
    r = np.random.randint(0,3)
    q = (q + p[r])/2.
    xlist.append(q[0])
    ylist.append(q[1])

plt.plot(xlist, ylist, 'k.', ms = 0.5)
#plt.axis('off')
plt.show()

```

```python
p = np.array(([0,0], [1,0],[0.5,0.5]))


def tri(p, count = 0, iterate = 5):
    if count < iterate:
        count += 1
        for i in range(len(p)):
            corner = [p[i]]
            for j in range(len(p)):
                x1 = mid(p[i], p[j])
                corner.append(x1)
                z = list(zip(p[i], x1))
                plt.plot(z[0], z[1], 'k')
            tri(corner, count)

tri(p)
```

```python
from turtle import *

def snowflake(lengthSide, levels):
    if levels == 0:
        forward(lengthSide)
        return
    lengthSide /= 3.5
    snowflake(lengthSide, levels-1)
    left(60)
    snowflake(lengthSide, levels-1)
    right(120)
    snowflake(lengthSide, levels-1)
    left(60)
    snowflake(lengthSide, levels-1)

def full_snowflake(lengthside, levels):
    for i in range(3):
        snowflake(lengthside, levels)
        right(120)

if __name__ == "__main__":
    speed(0)
    length = 500.0
    penup()
    backward(length/2.0)
    pendown()
    full_snowflake(length, 3)
    mainloop()
```
