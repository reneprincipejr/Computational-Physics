```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
L = 500
box = np.zeros([2*L,2*L])
box[L][L] = 1
for i in range(2*L): box[i][-1] = box[i][0] = box[0][i] = box[-1][i] = 2 
    
#0 = unoccupied
#1 = occupied
#2 = vacant
```

```python
plt.figure(figsize=(9,9))
plt.imshow(box,cmap = cm.gray_r)
plt.show()
```

```python
def jump(i,j):
    directions = [[0,1], [0,-1], [1,0], [-1,0]]
    choice = np.random.randint(0,4)
    
    i += directions[choice][0]
    j += directions[choice][1]
    
    return i, j
```

```python
p = 0.59275
count = 1
```

```python
x = y = 0
for i in range(1000000):
    r = np.random.random()

    x,y = jump(x,y)
    
    if abs(L+x) < 2*L and abs(L+y) < 2*L:
        if box[L + x][L + y] == 0:
            if r < p: 
                box[L + x][L + y] = 1
                count +=1
            else: box[L + x][L + y] = 2
        else: 
            continue
    
print(count)
plt.figure(figsize=(9,9))
plt.imshow(box,cmap =cm.gray_r)
plt.show()
```
