```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
num = 100

a = [2,3,4]

x = np.linspace(0,5,num)

integrand = np.zeros(num)

for i in a:
    power_x = i-1
    for i in range (np.size(x)):
        integrand[i] = x[i] ** power_x * np.exp(-x[i])
    
    plt.plot(x,integrand)
    
plt.show()
```
