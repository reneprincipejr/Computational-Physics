```python
from numpy import euler_gamma as y
from numpy import exp, linspace
import matplotlib.pyplot as plt
import math
```

Using the Weierstrass definition of the gamma function

```python
def gamma(z):
    N = 1000000
    f = 1
    for n in range(1, N+1):
        f*=(1+z/n)**(-1)*exp(z/n)
    return exp(-y*z)/z*f
```

```python
gamma(10)
```

Plotting the Gamma Function:

```python
z = linspace(-5,5,1000)
plt.plot(z,gamma(z))
plt.ylim(-5,5)
plt.ylabel("Gamma Value")
plt.xlabel("x Values")
plt.title("Plot of Gamma Function")
plt.show()
```

It can be seen that the plot described here is shifted to the right by one unit. This is because the plot shown in Arfken's book is $\Gamma$(x+1) but the plotted function here is $\Gamma$(x)

Comparing the values of the gamma function to the factorial from previous lesson:

```python
def factorial(z):
    if z==1:
        return 1
    else:
        return z*factorial(z-1)
```

```python
def error_factorial(x):
    k = 2
    er_lis = []
    while k <x:
        e = abs(gamma(k) - factorial(k-1))
        er_lis.append(e)
        k +=1
    x_ax = linspace(2,(x-1),(x-2))
    plt.plot(x_ax, er_lis)
    plt.semilogy()
    plt.ylabel("Logarithm of Error")
    plt.xlabel("x Values")
    plt.title("Error in comparison to Factorial")
    plt.show()
```

```python
error_factorial(100)
```

Logarithm was used in order to visualize more the increase of error. This rising of error can be attributed to the piling difference from the smaller values as the gamma function employs recursion.

```python
def error_math(x):
    k = 2
    er_lis = []
    while k <x:
        e = abs(gamma(k) - math.gamma(k))
        er_lis.append(e)
        k +=1
    x_ax = linspace(2,(x-1),(x-2))
    plt.plot(x_ax, er_lis)
    plt.semilogy()
    plt.ylabel("Logarithm of Error")
    plt.xlabel("x Values")
    plt.title("Error in comparison to Math Function")
    plt.show()
```
