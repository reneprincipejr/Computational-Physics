```python
from numpy import array,arange
import matplotlib.pyplot as plt

alpha = 1
beta = 0.5
gamma = 0.5
delta = 2

def f(r,t):
    x = r[0]
    y = r[1]
    fx = alpha*x - beta*x*y
    fy = gamma*x*y - delta*y
    return array([fx,fy],float)

a = 0.0
b = 30.0
N = 1000
h = (b-a)/N

tpoints = arange(a,b,h)
xpoints = []
ypoints = []

r = array([2.0,2.0],float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    r += (k1+2*k2+2*k3+k4)/6

plt.plot(tpoints,xpoints,label='rabits')
plt.plot(tpoints,ypoints,label='foxes')
plt.xlabel("t")
plt.ylabel('population in 1000')
plt.legend()
plt.show()
```

The Lotka-Volterra Model describes the simplest case of a predator-prey relationship. The cyclical behaviour of Rabits and Foxes is evidents, and the Fox population increase tracked the peaks of the Rabit population. This means that further increase in the number of Foxes will compromise the amount of prey, maybe due to competition. At some point, the continuous decline of the number of rabits will cause a decline in the number of foxes. Then, the continuous decline on the number of foxes, will at some point, cause an increase in the rabit population, and so on and so forth. However, this model is too simplistic since it doesn't take into account the complexity of a realistic situation, such as the randomness of the predator-prey encounters through a homogeneous environment.

Source: M. Beals, L. Gross, and S. Harrell, PREDATOR-PREY DYNAMICS: LOTKA-VOLTERRA, The Institute of Environmental Modeling, The University of Tennessee, Knoxville (1999).

According to Sharov (1996), This model has no asymptotic stability, it does not converge to an attractor (does not "forget" initial conditions).

"This figure shows relative changes in prey predator density for both initial conditions. Trajectories are closed lines.
The model of Lotka and Volterra is not very realistic. It does not consider any competition among prey or predators. As a result, prey population may grow infinitely without any resource limits. Predators have no saturation: their consumption rate is unlimited. The rate of prey consumption is proportional to prey density. Thus, it is not surprising that model behavior is unnatural showing no asymptotic stability. However numerous modifications of this model exist which make it more realistic."
